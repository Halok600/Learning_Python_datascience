```python
cols = list(df_bs.columns)
cols.remove('Critic rating')
insert_pos = cols.index('User Rating')+1
cols.insert(insert_pos,'Critic rating')
df_bs = df_bs[cols]
df_bs.head()

```

```python
df_bs['Avg Rating'] = df_bs['User Rating'] + df_bs['Critic rating']/2
df_bs['Avg Rating'] = df_bs['Avg Rating'].round(2)
df_bs.head()
```
